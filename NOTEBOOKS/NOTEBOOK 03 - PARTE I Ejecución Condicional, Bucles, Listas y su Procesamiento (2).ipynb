## **1\. Valores Booleanos**

### 1.1. **Teor칤a**

En Python, los valores booleanos son un tipo de dato que solo pueden tener dos valores: `True` (verdadero) y `False` (falso). Estos valores son fundamentales para el control de flujo en los programas, permitiendo que se tomen decisiones basadas en condiciones.

- `True` y `False` son palabras clave en Python.
- Se usan principalmente en expresiones condicionales y bucles.

#### **Operadores de comparaci칩n**: Estos operadores devuelven un valor booleano como resultado de comparar dos valores.

- `==` (igual a)
- `!=` (diferente de)
- `<`, `>`, `<=`, `>=` (menor, mayor, menor o igual, mayor o igual)

### 1.2. **Ejemplo de C칩digo**

````python
# Ejemplo de valores booleanos
x = 10
y = 20

# Operaci칩n de comparaci칩n
es_mayor = x > y  # False porque 10 no es mayor que 20
es_igual = x == y  # False porque 10 no es igual a 20

print(f"쮼s {x} mayor que {y}? {es_mayor}")
print(f"쮼s {x} igual a {y}? {es_igual}")
````

### **Caso: Suma de numeros del 1 a la cantidad que defina el usuario.**

Vamos a suponer que queremos calcular la suma de los n칰meros del 1 a la catidad que deficna el suario. Este caso es 칰til para entender c칩mo funciona el bucle `for`, ya que iteramos sobre una secuencia de n칰meros y realizamos una operaci칩n en cada iteraci칩n.

````python
#Respuesta
numusuario=int(input("ingrese un numero hasta al que se va a realizar la suma")) 
suma = 0
for i in range (1, numusuario+1):
   print (f" numero {i}")
   suma = suma + i
   print (f"la suma total es de {suma} ")
````

### 

- **Bucle** `while`:

````python
# Ejemplo de bucle while
num=int(input("ingrese un numero hasta al que se va a realizar la suma")) 
contador = 1 
suma = 0
while contador < num + 1:
    suma += contador # suma = suma + contador
    print (f"contador : {contador}")
    contador += 1
print ("la suma total es :", suma)
````
